**B. COMBINE AND REDUCING REELATED COLUMNS BY AVERAGING THEM**

```python

i = df['goalkeeping_diving'] + df['goalkeeping_handling'] + df['goalkeeping_kicking'] + df['goalkeeping_positioning'] + df['goalkeeping_reflexes']
df['GK'] = i/5
i = ['goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes']
df = df.drop(i, axis=1)

i = df['defending_sliding_tackle'] + df['defending_standing_tackle'] + df['defending_marking']
df['Defending'] = i/3
i = ['defending_sliding_tackle', 'defending_standing_tackle', 'defending_marking']
df = df.drop(i, axis=1)

i = df['mentality_composure'] + df['mentality_penalties'] + df['mentality_vision'] + df['mentality_positioning'] + df['mentality_interceptions'] + df['mentality_aggression']
df['Mentality'] = i/6
i = ['mentality_composure', 'mentality_penalties', 'mentality_vision', 'mentality_positioning', 'mentality_interceptions', 'mentality_aggression']
df = df.drop(i, axis=1)

i =df['power_long_shots'] = df['power_strength'] + df['power_stamina'] + df['power_jumping'] + df['power_shot_power']
df['Power'] = i/5
i = ['power_shot_power', 'power_jumping', 'power_stamina', 'power_strength','power_long_shots']
df = df.drop(i, axis=1)

i = df['movement_acceleration'] + df['movement_sprint_speed'] + df['movement_agility'] + df['movement_reactions'] + df['movement_balance']
df['Movement'] = i/5
i = ['movement_acceleration', 'movement_sprint_speed', 'movement_agility', 'movement_reactions','movement_balance']
df = df.drop(i, axis=1)
```

# **NUMBER TWO**

**FEATURES THAT SIGNIFICANTLY CORRELATE TO THE OVERAL**

```python
#selecting features
from sklearn.model_selection import train_test_split
train_data, test_data= train_test_split(df,test_size=0.2)
from sklearn.linear_model import LinearRegression
target_feature='overall'
#finding features that arecorrelated to the overall column
feature_corr=train_data.corr(method='pearson')[target_feature]
feature_corr=feature_corr.sort_values(ascending=False)
#print thetop ten correlations with the target value
print(feature_corr[1:21])
corr_matrix = df.corr()
corr_matrix['overall'].sort_values(ascending=False)
```

# **NUMBER THREE**
**CREATING AND TRAINNING A SUITABLE MACHINE LEARNING MODEL**

#copying top 20 relavent features to be used by model
features=feature_corr[1:11].index.tolist()
print(features)

```python
features=feature_corr[1:11].index.tolist()
print(features)
from sklearn.model_selection import train_test_split
train_data, test_data= train_test_split(df,test_size=0.2)
from sklearn.linear_model import LinearRegression
#training the model
x_train=train_data[features]
y_train=train_data[target_feature]

#replace all empty cells with zero
x_train.fillna(0,inplace=True)

#using the LinearRegression method to build the model
model=LinearRegression().fit(x_train,y_train)
#print score
print("Score:"+str(model.score(x_train,y_train)))
```

**TESTING THE MODEL USING THE 25% OF THE FIFA 20 DATASET**

```python
#testing the model
#sort test data first
test_data=test_data.sort_values([target_feature], ascending=False)
x_test=test_data[features]
x_test.fillna(0,inplace=True)
y_test=test_data[target_feature]

#start predicting
y_predict=model.predict(x_test)
#add new column called predicted 
test_data['predicted']=y_predict
rating=((y_predict-y_test)/y_test*100)
#add anew column called accuracy
test_data['difference']=rating
test_data[["short_name","overall","predicted","difference"]]
```
